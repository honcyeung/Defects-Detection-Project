```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import torch
import os
from torchvision import datasets, transforms, models
from skimage import io, transform
import matplotlib.pyplot as plt
from torch import nn, optim
from time import time
import torchvision
from IPython.display import Image
from IPython.core.display import HTML 
```

## 1. Image Data Preparation

```python
# parameters for architecture

# input_shape
num_classes = 6

# parameters for training
batch_size = 32
num_epochs = 20

# load csv fils containing image information
train = pd.read_csv('train_database.csv', sep = ";")

train_image = []

# load image files

for i in range(train.shape[0]):
    img = io.imread(os.path.join('Database/', train['Image'][i])) 
    #img = image.img_to_array(img)
    print(img.shape)
    img = img/255
    train_image.append(img)

X = np.array(train_image) 
y = train['Class'].values

x_train, x_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.1)
```

```python
# get the number of images

X.shape[0]
```

```python
# resize and transform each image to tensor

train_trainsforms = transforms.Compose([transforms.Resize((200,200)),
                                            transforms.ToTensor(),
                                            transforms.Normalize((0.5,), (0.5,)),])
    
test_trainsforms = transforms.Compose([transforms.Resize((200,200)),
                                           transforms.ToTensor(),
                                           transforms.Normalize((0.5,), (0.5,)),])
```

## 2. Data Preparation

```python
# get images for training and testing the programme

train_data = datasets.ImageFolder('Data_train/',transform = train_trainsforms)
test_data = datasets.ImageFolder('Data_test/',transform = test_trainsforms)
```

The number of image for training and testing in each category has increased to 50 respectively in order to increase the accuracy of the network. 

```python
# number of training sets

num_train = len(train_data)                               
indices = list(range(num_train))   
```

```python
# number of training sets

num_test = len(test_data)                               
indices = list(range(num_test))
```

```python
# the loader with load each time 64 images for one training itteration

train_loader =  torch.utils.data.DataLoader(train_data, batch_size = 64, shuffle = True)
test_loader =  torch.utils.data.DataLoader(test_data, batch_size = 64, shuffle = True)
```

## 3. Build the Network

```python
# input size

input_size = 40000

# size of hidden layers

hidden_sizes = [600, 450]
output_size = 10

# neural network

model = nn.Sequential(nn.Linear(input_size, hidden_sizes[0]),
                      nn.ReLU(),
                      nn.Linear(hidden_sizes[0], hidden_sizes[1]),
                      nn.ReLU(),
                      nn.Linear(hidden_sizes[1], output_size),
                      nn.LogSoftmax(dim = 1))
print(model)
```

```python
# define negative log likelihood loss

criterion = nn.NLLLoss()
images, labels = next(iter(train_loader))
images = images[:,0,:,:]

# view(a, b) will return a new tensor with the same data as `images` with size `(a, b)`.
images = images.view(images.shape[0], -1)

# log probabilities, the output of our neural network model
logps = model(images) 

# calculate the NLL loss
loss = criterion(logps, labels) 
```

```python
# get the dimension of image

images.shape
```

```python
print('Before backward pass: \n', model[0].weight.grad)

loss.backward()

print('After backward pass: \n', model[0].weight.grad)
```

```python
# Stochastic Gradient Descent optimizer

optimizer = optim.SGD(model.parameters(), lr = 0.003, momentum = 0.9)
time0 = time()

# number of feedforward and backpropagation iterations that we will do to train the network
epochs = 30

for e in range(epochs):
    
    running_loss = 0
    
    # iterate through the image dataset
    for images, labels in train_loader:
        
        # flatten images into a vector
        images = images[:,0,:,:]
        images = images.view(images.shape[0], -1)
        
        # training pass
        optimizer.zero_grad()
        
        output = model(images)
        loss = criterion(output, labels)
        
        # model learnt by backpropagating
        loss.backward()
        
        # optimize its weights
        optimizer.step()
        
        running_loss += loss.item()
    
    else:
        print("Epoch {} - Training loss: {}".format(e, running_loss/len(train_loader)))

print("\nTraining Time (in minutes) =", (time() - time0)/60)    
```

```python
for images, labels in train_loader:
    print(images.shape)
```

```python
correct_count, all_count = 0, 0

for images, labels in test_loader:
    images = images[:,0,:,:]
    for i in range(len(labels)):
        img = images[i].view(1, 40000) 
        with torch.no_grad():
            logps = model(img)
            
        # get the scores obtained for each class
        ps = torch.exp(logps)
        probab = list(ps.numpy()[0])
        
        # choose the one having the max score and compare it to the real label
        pred_label = probab.index(max(probab))
        true_label = labels.numpy()[i]
        
        if(true_label == pred_label):
            correct_count += 1
        
        all_count += 1
        
print("Number of Images Tested =", all_count)
print("\nModel Accuracy =", correct_count/all_count)
```

The number of tested images icreases to 300, 50 in each category. The accuracy has surged from around 20% to around 49%. We can foresee that accuracy and the number of tested image are directly proportioanl to each other.
